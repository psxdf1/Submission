```python
import collections.abc
#reframed needs the four following aliases to be done manually. SHannanigans between python3.9 & 3.10
collections.Iterable = collections.abc.Iterable
collections.Mapping = collections.abc.Mapping
collections.MutableSet = collections.abc.MutableSet
collections.MutableMapping = collections.abc.MutableMapping
import reframed
```

```python
# load model and convert to reframed format

from reframed import load_cbmodel
AG = load_cbmodel('../../../data/AG.xml', flavor='fbc2')
GD = load_cbmodel('../../../data/GD.xml', flavor='fbc2')
```

```python
# COntrol.
import micom as mc
com = mc.load_pickle("../../../data/community.pickle")
com.cooperative_tradeoff(fraction=0.96,pfba=True,fluxes=False)
```

```python
# Create reframed community format-
co = reframed.Community("LiuCom", [AG,GD])
```

```python
from reframed.solvers import get_default_solver
print(get_default_solver())
# co.merged_model
```
